importing the dependencies or files


```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

data collection and data processing

```python
#loading the data set to a pandas dataframe
sonar_data = pd.read_csv('/content/sonar data.csv', header=None)
```

```python
sonar_data.head()
```

```python
#number of rows and coloumns
sonar_data.shape
```

```python
sonar_data.describe() #describ --> statical measures of the data
```

```python
sonar_data[60].value_counts()
```

M---> mine
R---> rock

```python
sonar_data.groupby(60).mean()
```

```python
# saparating data and lables
X = sonar_data.drop(columns=60, axis=1)
Y = sonar_data[60]
```

```python
print(X)
print(Y)
```

training and test data

```python
X_train , X_test , Y_train , Y_test = train_test_split(X,Y,test_size=0.1,stratify=Y,random_state=1)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
print (X_train)
print (Y_train)
```

modal training ---> logistic regression

```python
model = LogisticRegression()
```

```python
#tranining the logistic regression model with traning data
model.fit(X_train,Y_train)
```

model evaluation

```python
#accuracy on traning data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print('accuracy on traning data :',training_data_accuracy)
```

```python
#accuracy on traning data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print('accuracy on test data :',test_data_accuracy)
```
